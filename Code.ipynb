```python
rows = img.shape[0]
cols = img.shape[1]
```

```python
land = [255, 38, 0, 255]

def isLand(pixel):
    # pixel is of form [r, g, b, 255]
    for i in range(4):
        if pixel[i] != land[i]:
            return False
    return True
```

```python
land_count = 0

for r in range(rows):
    for c in range(cols):
        current_pixel = img[r][c]
        if isLand(current_pixel):
            land_count += 1
            original[r][c] = [255, 0, 253, 255]
```

```python
plt.figure(figsize=(10, 8))
plt.imshow(original)
```

```python
# land_ratio
land_count / (rows * cols)
```
